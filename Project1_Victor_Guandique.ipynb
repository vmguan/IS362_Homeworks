## Perform analysis to compare the delays between Alaska and AM West

```python
# Import pandas and numpy
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

# Read project1.csv 
airlines_status = pd.read_csv('/Users/vguan/Documents/SPS/IS362/Week 4/project1.csv')

# Shows the information from project.csv
airlines_status.head()
```

## Shows the Total of Flights per airline for every destination

```python
airlines_status['Alaska Total'] = airlines_status['Alaska On Time'] + airlines_status['Alaska Delayed']
airlines_status['AM West Total'] = airlines_status['AM West On Time'] + airlines_status['AM West Delayed']
airlines_status.head()
```

## Shows the percentage of flights delayed per airline for every destination

```python
airlines_status['Alaska Percentage Delayed'] = (airlines_status['Alaska Delayed'] / airlines_status['Alaska Total']) * 100
airlines_status['AM West Percentage Delayed'] = (airlines_status['AM West Delayed'] / airlines_status['AM West Total']) * 100
airlines_status.head()
```

## Shows which airline has the higher amount of delays

```python
status = [airlines_status['Alaska Percentage Delayed'] >= airlines_status['AM West Percentage Delayed'], 
            airlines_status['Alaska Percentage Delayed'] <= airlines_status['AM West Percentage Delayed']] 

options = ['Alaska', 'AM West']

airlines_status['Higher Delays'] = np.select(status, options)
airlines_status.head()
```

## The analysis shows that AM West has the highest number of delays per destination compare to Alaska
