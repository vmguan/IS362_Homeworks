# IS 362 Week 14 Assignment

### Youtube is probably the biggest video sharing platform. It offers videos for all users, going from music videos to personal videos and more. It has a recommender system that many people consider very effective.

## -  Target Users

### The main site is targeting users 17 and up but it also has an option for younger kids. Youtube is targeting the users not just offering similar videos to the one users watch but also by providing advertising that in many cases is related to the content the users watch.

## -  Key Goals

### Their main goal is to improve the user experience by suggesting similar content to the one user's. Another goal is to provide revenue to the company. 

## - How Can They Accomplish Those Goals

### There are many ways youtube can accomplish these goals. First they collect data from customers browsing history to make suggestions in the future. They also use other options as likes or dislikes, customers adding videos to favorites or subscribing to specific channels. They used all this information not just to suggest videos that customers might like but also to show customized advertising when customers are watching videos.
