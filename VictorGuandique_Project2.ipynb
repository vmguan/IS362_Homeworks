# Tidying Up Data

```python
# Import pandas and numpy
import pandas as pd
import numpy as np
import datetime
from datetime import datetime, date
```

## Read the File with the First Dataset

```python
students_results = pd.read_csv('/Users/vguan/Documents/SPS/IS362/Week 6/student_results.csv')
students_results.head(10)
```

## Checking for duplicated data

```python
students_results.duplicated()
```

## Splitting sex and age in two columns and removing 'sex and age' and 'phone' columns

```python
students_results[['sex','age']]=students_results['sex and age'].str.split('_', n = 1, expand = True)
students_results.drop(['sex and age','phone'], inplace = True, axis = 1)
students_results.head(10).sort_values(by=['id'])
```

## Calculate the average of all tests

```python
students_results['Average'] = students_results.iloc[:, 3:5].mean(axis = 1).round(2)
students_results.head(10).sort_values(by=['id'])
```

## As we can see on the Average column 4 students got a higher average on test 2 compare to test 1

```python
students_results.pivot(index='name', columns = ['test number'], values = 'Average')
```
