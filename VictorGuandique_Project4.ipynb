## Calculating Odor Accuracy Predicting Poisonous.

```python
data_odor=data_num[['edible_or_poisonous', 'odor']]
data_odor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

y = data_odor['edible_or_poisonous']

X = data_odor.drop(["edible_or_poisonous"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size =0.2)
```

```python
model = DecisionTreeClassifier()

model.fit(X_train, y_train)

odor_prediction = model.predict(X_test)

odor_score = accuracy_score(y_test, odor_prediction)

total_odor_percentage = odor_score * 100
print('Odor Accuracy Predicting if Poisonous: '+ str(total_odor_percentage))
```
